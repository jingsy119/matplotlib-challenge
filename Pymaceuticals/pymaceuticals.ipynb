### Analysis

- The drug of interest, Capomulin, resulted in smaller tumor volume at the end of the treatment period (45 days) as compared to other drugs, though Capomulin showed similar effectiveness compared to that of Ramicane.
- Generally, the tumor volume decreased as the treatment time increased for drug Capomulin. The smallest tumor volume of less than 41 mm3 was noted at around 34 days of treatment. 
- The tumor volume had a fairly strong positive correlation with the weight of the treated mice. That is, the heavier the mice's weight, the larger the tumor volume. 
- Overall, the data does not appear to have strong bias, as the number of mice tested for each drug was similar in size. Moreover, the sex of the mice should also not interfere with the test result, as there was a fairly even distribution of male and female mice.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results, on = "Mouse ID")

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mice_number = len(combined_data["Mouse ID"].value_counts())
mice_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_data = combined_data[combined_data[["Mouse ID", "Timepoint"]].duplicated() == True]
duplicated_data["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_rows = combined_data.groupby(["Mouse ID", "Timepoint"])["Mouse ID"].transform("size") > 1
duplicate_mouse = combined_data[duplicated_rows]
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_combined_data = combined_data.loc[(combined_data["Mouse ID"] != "g989"), :]
cleaned_combined_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mice_number = len(cleaned_combined_data["Mouse ID"].value_counts())
cleaned_mice_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_tumor_vol = cleaned_combined_data.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
median_tumor_vol = cleaned_combined_data.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
var_tumor_vol = cleaned_combined_data.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
std_tumor_vol = cleaned_combined_data.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
sem_tumor_vol = cleaned_combined_data.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

stat_summ_tumor = pd.concat([mean_tumor_vol, median_tumor_vol, var_tumor_vol, std_tumor_vol, sem_tumor_vol], axis = 1)
stat_summ_tumor.columns = ["Mean Tumor Volume", "Median Tumor Volume", "Tumor Volume Variance", "Tumor Volume Std. Dev.", "Tumor Volume Std. Err."]
stat_summ_tumor
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
stat_summ_tumor_agg = cleaned_combined_data.groupby(["Drug Regimen"]).agg({
    "Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
stat_summ_tumor_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
cric_regimen = cleaned_combined_data.loc[(cleaned_combined_data["Drug Regimen"] == "Capomulin") | (cleaned_combined_data["Drug Regimen"] == "Ramicane") |
                                            (cleaned_combined_data["Drug Regimen"] == "Infubinol") | (cleaned_combined_data["Drug Regimen"] == "Ceftamin"), :]
max_time_df = pd.DataFrame(cric_regimen.groupby(["Mouse ID"])["Timepoint"].max()).reset_index(drop = False)
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_time_df_merge = pd.merge(max_time_df, cleaned_combined_data, on = ["Mouse ID", "Timepoint"], how = "left")
max_time_df_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treat_label = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

for i in range(len(treat_label)):
    tumor_vol_data.append(list(max_time_df_merge[(max_time_df_merge["Drug Regimen"] == treat_label[i])]["Tumor Volume (mm3)"].sort_values()))
print(f"The tumor volume data for each drug regimen are {tumor_vol_data}")

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles = []
iqr = []
upper_bound = []
lower_bound = []

for j in range(len(tumor_vol_data)):
    quartiles.append(np.percentile(tumor_vol_data[j], [25, 75]))
    iqr.append(quartiles[j][1] - quartiles[j][0])
    upper_bound.append(quartiles[j][1] + 1.5*iqr[j])
    lower_bound.append(quartiles[j][0] - 1.5*iqr[j])
    
print(f"The iqr of {treat_label[0]} is {round(iqr[0], 2)}. The iqr of {treat_label[1]} is {round(iqr[1], 2)}. The iqr of {treat_label[2]} is {round(iqr[2], 2)}. The iqr of {treat_label[3]} is {round(iqr[3], 2)}")   
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
        # add subset 
        # Determine outliers using upper and lower bounds
cap_subset = cleaned_combined_data.loc[(cleaned_combined_data["Drug Regimen"] == treat_label[0]), ["Tumor Volume (mm3)"]]
cap_outlier = cap_subset.loc[(cap_subset["Tumor Volume (mm3)"] < lower_bound[0])
                             | (cap_subset["Tumor Volume (mm3)"] > upper_bound[0])].reset_index(drop = True).squeeze() 

ram_subset = cleaned_combined_data.loc[(cleaned_combined_data["Drug Regimen"] == treat_label[1]), ["Tumor Volume (mm3)"]]
ram_outlier = ram_subset.loc[(ram_subset["Tumor Volume (mm3)"] < lower_bound[1])
                             | (ram_subset["Tumor Volume (mm3)"] > upper_bound[1])].reset_index(drop = True).squeeze() 

inf_subset = cleaned_combined_data.loc[(cleaned_combined_data["Drug Regimen"] == treat_label[2]), ["Tumor Volume (mm3)"]]
inf_outlier = inf_subset.loc[(inf_subset["Tumor Volume (mm3)"] < lower_bound[2])
                             | (inf_subset["Tumor Volume (mm3)"] > upper_bound[2])].reset_index(drop = True).squeeze() 

cef_subset = cleaned_combined_data.loc[(cleaned_combined_data["Drug Regimen"] == treat_label[1]), ["Tumor Volume (mm3)"]]
cef_outlier = cef_subset.loc[(cef_subset["Tumor Volume (mm3)"] < lower_bound[1])
                             | (cef_subset["Tumor Volume (mm3)"] > upper_bound[1])].reset_index(drop = True).squeeze() 

print(f"Capomulin's potential outliers: {cap_outlier}")
print(f"Ramicane's potential outliers: {ram_outlier}")
print(f"Infubinol's potential outliers: {round(inf_outlier, 4)}")
print(f"Ceftamin's potential outliers: {cef_outlier}")    

    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig, ax = plt.subplots()
ax.boxplot(tumor_vol_data, labels = treat_label, sym= "red")
ax.set_ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
mouse = "l509"
line_df = cleaned_combined_data.loc[(cleaned_combined_data["Mouse ID"] == mouse) & (cleaned_combined_data["Drug Regimen"] == "Capomulin"), :]
plt.plot(line_df["Timepoint"], line_df["Tumor Volume (mm3)"])
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse l509")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
scatter_df = cleaned_combined_data.loc[(cleaned_combined_data["Drug Regimen"] == "Capomulin"), :]
scatter_df_grp = pd.DataFrame(scatter_df.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean().reset_index(drop = False))

plt.scatter(scatter_df_grp["Weight (g)"], scatter_df_grp["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse average tumor volume over weight")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

from scipy.stats import linregress

x_values = scatter_df_grp["Weight (g)"]
y_values = scatter_df_grp["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")

plt.annotate(line_eq, (10, 1), fontsize=15, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse average tumor volume over weight")
print(f"The correlation between mouse weight and the average tumor volume is {round(rvalue, 2)}")
plt.show()
```
